```python

#to check mobile number 

def is_num(s):
    
    
    # 1) Then contains 6,7 or 8 or 9.
    # 2) Then contains 9 digits
    Pattern = re.compile("[6-9][0-9]{9}")
    

    if re.match(Pattern, s) and len(s)==10: 
        return s
    else :
        return '0000000000' 
```

```python
import string



def string_fx(a):
    invalidcharacters= list(string.punctuation)
    invalidcharacters.remove('.')
    invalidcharacters.remove('@')
    invalidcharacters.remove('_')

    if any(char in invalidcharacters for char in a):
        return "invalid"
    else:
        return a
```

```python
string_fx('saksham.kumar82@gmail.com')
```

```python
def email_set(h,g):
    c = g
    b = h.shape[0]
    for i in c:
        for j in range(b):
            h.loc[j,i]=string_fx(h.loc[j,i])
            
```
